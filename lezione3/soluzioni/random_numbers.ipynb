### Istogrammi di Distribuzioni Uniforme e Normale  

In questo esercizio lavoreremo con due tipi di distribuzioni statistiche:  
1. La **distribuzione uniforme**, dove ogni valore ha la stessa probabilità di occorrere.  
2. La **distribuzione normale (o gaussiana)**, che segue una forma a campana con i valori più probabili vicini alla media.  

Per ciascuna distribuzione, genereremo numeri casuali, creeremo un **istogramma delle frequenze relative** e sovrapporremo la curva teorica della distribuzione per verificarne la forma. Questo esercizio dimostra come combinare codice e teoria utilizzando celle di **markdown** e **codice** in un notebook.  


```python
# Importare i moduli necessari
import random
import matplotlib.pyplot as plt
import numpy as np

# Definire il numero di numeri casuali da generare
N = 1000

# Funzione per tracciare la curva teorica
def plot_theoretical_curve(x, y, label, color='red'):
    plt.plot(x, y, label=label, color=color, linewidth=2)
    plt.legend()
```

### Distribuzione Uniforme  

Una distribuzione uniforme è caratterizzata dal fatto che ogni valore in un intervallo ha la stessa probabilità di essere scelto.  

La sua funzione di densità di probabilità (PDF) è:  
$$
f(x) = 
\begin{cases} 
\frac{1}{b - a} & \text{se } x \in [a, b] \\
0 & \text{altrimenti}
\end{cases}
$$ 

Nel nostro caso, genereremo $N$ numeri casuali nell'intervallo \([0, 1]\). L'istogramma mostrerà le **frequenze relative**, e sovrapporremo una linea costante che rappresenta la PDF teorica.  


```python
# Generare N numeri casuali uniformi
uniform_random_numbers = [random.uniform(0, 1) for _ in range(N)]

# Calcolare le frequenze relative
bins = np.linspace(0, 1, 21)  # 20 intervalli
hist, bin_edges = np.histogram(uniform_random_numbers, bins=bins, density=True)

# Creare l'istogramma
plt.bar(bin_edges[:-1], hist, width=(bins[1] - bins[0]), edgecolor='black', color='skyblue', alpha=0.7, label="Frequenze relative")

# Aggiungere la curva teorica della PDF uniforme
x_uniform = np.linspace(0, 1, 100)
y_uniform = [1] * len(x_uniform)  # PDF uniforme è costante
plot_theoretical_curve(x_uniform, y_uniform, label="PDF teorica")

# Personalizzare il grafico
plt.title("Distribuzione Uniforme")
plt.xlabel("Valore")
plt.ylabel("Frequenza relativa")
plt.grid(True, linestyle='--', alpha=0.7)

# Salvare il grafico su file
plt.savefig("out/uniform_distribution_plot.png")

# Mostrare il grafico
plt.show()
```

### Distribuzione Normale  

La distribuzione normale, o gaussiana, è caratterizzata dalla sua forma a campana, centrata sulla media e con una deviazione standard che ne determina la larghezza.  

La sua funzione di densità di probabilità (PDF) è:  
$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

Dove:  
- $\mu$ è la media.  
- $\sigma$ è la deviazione standard.  

Nel nostro caso, genereremo 1000 numeri casuali con media $\mu = 0$ e deviazione standard $\sigma = 1$. L'istogramma mostrerà le **frequenze relative**, e sovrapporremo la curva della PDF teorica per confronto.  


```python
# Generare N numeri casuali normali con media 0 e deviazione standard 1
normal_random_numbers = [random.gauss(0, 1) for _ in range(N)]

# Calcolare le frequenze relative
bins = np.linspace(-4, 4, 21)  # 20 intervalli
hist, bin_edges = np.histogram(normal_random_numbers, bins=bins, density=True)

# Creare l'istogramma
plt.bar(bin_edges[:-1], hist, width=(bins[1] - bins[0]), edgecolor='black', color='lightgreen', alpha=0.7, label="Frequenze relative")

# Aggiungere la curva teorica della PDF normale
x_normal = np.linspace(-4, 4, 100)
y_normal = (1 / (np.sqrt(2 * np.pi))) * np.exp(-0.5 * x_normal**2)
plot_theoretical_curve(x_normal, y_normal, label="PDF teorica")

# Personalizzare il grafico
plt.title("Distribuzione Normale")
plt.xlabel("Valore")
plt.ylabel("Frequenza relativa")
plt.grid(True, linestyle='--', alpha=0.7)

# Salvare il grafico su file
plt.savefig("out/normal_distribution_plot.png")

# Mostrare il grafico
plt.show()
```
